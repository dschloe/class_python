```
import requests
from bs4 import BeautifulSoup

def crawl_contents(soup) :
    # 도서의 제목들을 담고 있는 리스트를 반환하세요.
    # print(soup)
    div = soup.find("ol", class_="row")
    # print(div)
    titles = []
    for li in div.find_all("li"):
      titles.append(li.find("h3").get_text())
    
    return titles

def main():
    custom_header = {'user-agent' : 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_14_4) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/73.0.3683.103 Safari/537.36'}
    
    # 주어진 url의 html을 담고 있는 BeautifulSoup 객체를 매개변수로 crawl_contents 함수를 호출하세요.
    # crawl_contents 함수의 반환값을 출력하세요.
    url = "http://books.toscrape.com/"
    req = requests.get(url, headers = custom_header)
    soup = BeautifulSoup(req.text, "html.parser")
    
    print(crawl_contents(soup))
    

if __name__ == "__main__":
    main()
```

## 실전 테스트 3

```
import requests
from bs4 import BeautifulSoup

def crawl_contents(soup) :
    # 명언들의 태그의 빈도를 담고 있는 딕셔너리를 반환하세요.
    div = soup.find_all("div", class_="quote")
    # print(div)
    tags_name = []
    for div in div:
      a_tag_list = div.find("div", class_="tags").find_all("a", class_="tag")
      for a in a_tag_list:
        tags_name.append(a.get_text())
    return tags_name

def main():
    custom_header = {'user-agent' : 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_14_4) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/73.0.3683.103 Safari/537.36'}
    
    # 주어진 url의 html을 담고 있는 BeautifulSoup 객체를 매개변수로 crawl_contents 함수를 호출하세요.
    # crawl_contents 함수의 반환값을 출력하세요.
    
    list_href = []
    for i in range(0, 10):
        url = "http://quotes.toscrape.com/page/" + str(i+1)
        list_href.append(url)
    
    tag_list = []
    for href in list_href :
        href_req = requests.get(href, headers = custom_header)
        href_soup = BeautifulSoup(href_req.text, "html.parser")
        tag_list.append(crawl_contents(href_soup))
  
    result = []
    for word in tag_list:
      result.extend(word)
    # print(result)

    word_count = {}
    for word in result:
      if word in word_count:
        word_count[word] +=1
      else:
        word_count[word] = 1
    sort_orders = sorted(word_count.items(), key=lambda x: x[1], reverse=True)

    for i in sort_orders:	    
      print(i[0], i[1])

if __name__ == "__main__":
    main()
```
